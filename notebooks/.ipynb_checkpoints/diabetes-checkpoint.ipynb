```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
import ripser
from ripser import Rips
import persim
```

```python
df = pd.read_csv("../data/diabetes.csv")
df.head()
```

```python
# Summary statistics of the dataset
df.describe(include='all')
```

```python
# Visualize with histograms
for col in df.keys(): 
    df[col].hist(bins=20)
    plt.xlabel(col)
    plt.title(col)
    plt.show(block=True)
```

```python
# BASE MODEL (no feature engineering)

# Dependent var
y = df["Outcome"]

# Independent vars
X = df.drop("Outcome", axis=1)

# Split into test and training dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=17)
```

```python
def print_classifier_performance(name, y_pred, y_test):
    print(name)
    print(f"Accuracy: {round(accuracy_score(y_pred, y_test), 4)}")
    print(f"Recall: {round(recall_score(y_pred,y_test),4)}")
    print(f"Precision: {round(precision_score(y_pred,y_test), 4)}")
    print(f"F1: {round(f1_score(y_pred,y_test), 4)}")
    print(f"Auc: {round(roc_auc_score(y_pred,y_test), 4)}")
```

```python
# Random forest

# Train

rf_model = RandomForestClassifier(random_state=46).fit(X_train, y_train)

# Predict

rf_pred = rf_model.predict(X_test)

print_classifier_performance("Random Forest Classifier", rf_pred, y_test)
```

```python
# Linear regression

lr_model = LogisticRegression(random_state=46).fit(X_train, y_train)

lr_pred = lr_model.predict(X_test)

print_classifier_performance("Logistic Regression", lr_pred, y_test)
```

```python
# Linear regression with preprocessing (to avoid iterations limit error)

scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

lr_processed_model = LogisticRegression(random_state=46).fit(X_train_scaled, y_train)

lr_processed_pred = lr_processed_model.predict(X_test_scaled)

print_classifier_performance("Logistic Regression", lr_processed_pred, y_test)
```

```python
# TDA feature extraction

def diagram_sizes(dgms):
    return ", ".join([f"|$H_{i}$|={len(d)}" for i, d in enumerate(dgms)])

# Generate persistence diagrams
dgm = ripser.ripser(X_train)['dgms']
persim.plot_diagrams(
    dgm,
    show=True,
    title=f"Training Data Persistence Diagram \n{diagram_sizes(dgm)}"
)

dgm_scaled_data = ripser.ripser(X_train_scaled)['dgms']
persim.plot_diagrams(
    dgm_scaled_data,
    show=True,
    title=f"Scaled Training Data Persistence Diagram \n{diagram_sizes(dgm)}"
)
```

```python
# Extract Betti numbers
betti_numbers = [len(d) for d in dgm_scaled_data]
print(betti_numbers)
```
